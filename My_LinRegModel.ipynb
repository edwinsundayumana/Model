```python
import numpy as np
import matplotlib.pyplot as pt
from sklearn.linear_model import LinearRegression
```

```python
# create a list
x = np.array([1,10,20,40,50,70,80,90,120])
y = np.array([3,20,90,110,130,170,150,200,260])
```

```python
# create the linear regModel
linreg = LinearRegression()

# reshape x
x = x.reshape(-1,1)
linreg.fit(x,y)

# predict the value of y
y_pred = linreg.predict(x)

# plot the scatter plot
pt.scatter(x,y)
pt.plot(x,y_pred, color = 'red')
pt.xlabel('Rain', size = '10')
pt.ylabel('Sales', size = '10')
pt.title('A Simple Model')
# pt.savefig('L_regModel.png', dpi = 200)
pt.show()
```
